# So lets now define our sigmoid function

```python
def sigmoid(z):
    return 1.0/ (1+ np.exp(-z))
```

```python
sigmoid(X)
```

# Making Predictions

using our knowledge of our sigmoid function, we can now make predictions and probabilities. we can write a prediction function. This function would return the probability of our observation being true. we call this class 1 and its notation is (class = 1). as the probability gets closer to 1, our model is more confident that the observation is in class 1

```python
weights = []

weights = [[0.5, 0.4, 0.3, 0.7, 0.9, 0.85, 0.7, 0.65, 0.6, 0.2, 0.15, 0.14, 0.16, 0.18, 0.20]]
```

```python
weights
```

```python
def predict(features, weights):
    z = np.dot(features, weights)
    
    return sigmoid(z) #as defined earlier
```

```python
def cost_function(x, y):
    observations  = len(X)
    
    predictions = predict(X, weights.T)
    
    class1_cost = -X*np.log(predictions)
    
    class2_cost = (1-X)*np.log(1-predictions)
    #Take the sum of both costs
    cost = class1_cost - class2_cost
    #Take the average cost
    cost = cost.sum() / observations
    return cost
```
